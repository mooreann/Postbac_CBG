```python
overlap_mac_ids = set(neurod_mac[0]) & set(topmed_mac[0])
overlap_maf_ids = set(neurod_maf[0]) & set(topmed_maf[0])
```

```python
overlap_mac_df = neurod_mac.loc[neurod_mac[0].isin(overlap_mac_ids)]
overlap_mac_list = overlap_mac_df[1].tolist()
overlap_maf_df = neurod_maf.loc[neurod_maf[0].isin(overlap_maf_ids)]
overlap_maf_list = overlap_maf_df[1].tolist()
```

```python
mac = pd.Series(overlap_mac_list, name='minor allele count')
maf = pd.Series(overlap_maf_list, name='minor allele frequencies')
```

```python
print(maf)
```

```python
f4 = plt.figure(figsize=(12,25))

#mac plot
plt.subplot(1,2,1)
sns.distplot(mac)
plt.title('MAC of overlap')

plt.subplot(1,2,2)
sns.distplot(maf)
plt.title('MAF of overlap')

plt.show()

f4.savefig("MAC,MAF_dist.png")
```

```python
overlap_mac_df2 = topmed_mac.loc[topmed_mac[0].isin(overlap_mac_ids)]
overlap_mac_list2 = overlap_mac_df2[1].tolist()
overlap_maf_df2 = topmed_maf.loc[topmed_maf[0].isin(overlap_maf_ids)]
overlap_maf_list2 = overlap_maf_df2[1].tolist()

mac2 = pd.Series(overlap_mac_list2, name='minor allele count')
maf2 = pd.Series(overlap_maf_list2, name='minor allele frequencies')
```

```python
f5 = plt.figure(figsize=(12,25))

#mac plot
plt.subplot(1,2,1)
sns.distplot(mac2)
plt.title('MAC of overlap')

plt.subplot(1,2,2)
sns.distplot(maf2)
plt.title('MAF of overlap')

plt.show()

f5.savefig("MAC,MAF_dist2.png")
```
